En esta publicación vamos a demostrar la siguiente igualdad termodinámica mostrada en la Ecuación (1) para la velocidad del sonido ($c$), desde cero, es decir, a partir de ecuaciones fundamentales paso a paso, y escribo una breve introducción para aquellos lectores de Quora que no estén familiarizados con la termodinámica clásica, pero aun así tienen curiosidad de saber un poco más sobre esta disciplina.

$\displaystyle c^2:= \left ( \frac{\partial P}{\partial \rho} \right)_{\! S} = \frac{C_P}{C_V}  \left ( \frac{\partial P}{\partial \rho} \right)_{\! T} \tag{1}$

donde $P$ es la presión absoluta, $T$ es la temperatura absoluta, $\rho$ es la densidad, $S$ es la entropía, y $C_P, C_V$ son los calores específicos a presión y volumen constante, respectivamente. La igualdad (1) es verdadera para gases reales (no sólo para gases ideales).

__1. Introducción:__

Para empezar, tenemos la siguiente notación "especial" para las derivadas parciales:

$\displaystyle \left ( \frac{\partial z}{\partial x} \right)_{\! y}  \tag{2}$

La expresión (2) indica "derivada parcial de $z$ respecto a $x$ con $y$ constante". Utilizaremos únicamente las siguientes herramientas matemáticas del cálculo multivariable:

(i) Regla de la cadena a $w$ constante:

$\displaystyle \left ( \frac{\partial z}{\partial x} \right)_{\! w} = \left ( \frac{\partial z}{\partial y} \right)_{\! w} \cdot \left ( \frac{\partial y}{\partial x} \right)_{\! w}  \tag{3a}$

(ii) El recíproco de la derivada parcial:

$\displaystyle \left ( \frac{\partial x}{\partial z} \right)_{\! y} = \frac{1}{\displaystyle \left ( \frac{ \partial z }{ \partial x} \right)_{\! y}} \tag{3b}$

(iii) Diferencial exacta de $z$ como función de $x,y$ (ya veremos más adelante por qué usamos dos variables):

$\displaystyle z = Z(x,y) ~~~\Rightarrow~~~ dz =  \left ( \frac{\partial z}{\partial x} \right)_{\! y} dx + \left ( \frac{\partial z}{\partial y} \right)_{\! x} dy \tag{3c}$

(iv) La relación cíclica entre las variables $x,y,z$, que es la relación entre las derivadas parciales cuando la superficie en $\mathbb{R}^3$ se escribe como $x = X(y,z)$, $y = Y(x,z)$ ó $z = Z(x,y)$, es una consecuencia directa de la Ecuación (3c) aplicada dos veces, esta identidad la demostraremos más adelante:

$\displaystyle \left ( \frac{\partial z}{\partial x} \right)_{\! y}  \left ( \frac{\partial y}{\partial z} \right)_{\! x} \left ( \frac{\partial x}{\partial y} \right)_{\! z} = -1 ~~~\Rightarrow~~~ \left ( \frac{\partial z}{\partial x} \right)_{\! y} = - \left ( \frac{\partial z}{\partial y} \right)_{\! x} \left ( \frac{\partial y}{\partial x} \right)_{\! z}\tag{3d}$    

De la termodinámica clásica, primero utilizaremos el concepto más importante, el __Concepto de Estado__: 

> Si fuera posible conocer las masas, velocidades, posiciones y todos los modos de movimiento de todas las partículas constituyentes de un sistema, este conocimiento serviría para describir el estado microscópico del sistema que a su vez, determinaría todas las propiedades del sistema. Sin embargo, en ausencia del conocimiento detallado que se requiere para determinar el estado microscópico del sistema, la termodinámica comienza con una consideración de las propiedades del sistema que una vez determinadas, definen el estado macroscópico del sistema, es decir, cuando todas las propiedades son fijas, entonces el estado macroscópico del sistema es fijo, de hecho, se encuentra que cuando los valores de un pequeño número de propiedades son fijos, los valores de todas las demás también son fijas. Cuando se considera un sistema simple como una cantidad dada de una sustancia de composición determinada, __la fijación de los valores de DOS de las propiedades__, determina los valores de todas las demás. Por lo tanto, sólo dos propiedades o variables son independientes y todas las demás propiedades son variables dependientes. El estado termodinámico de un sistema simple se fija únicamente cuando los valores de las dos variables independientes son fijos. Las propiedades más susceptibles de control son la la temperatura ($T$) y presión ($P$) del sistema. Cuando $T$ y $P$ son fijos, el estado del sistema simple es fijo y todas las demás propiedades tienen valores únicos correspondientes a este estado $^{\mathbf{[1]}}$. 

Esto se conoce también como el teorema de Duhem (1886):

> Una vez la composición del sistema está definido, sólo DOS variables son suficientes para especificar el estado del sistema.

Más adelante, Gibbs (1906) $^{\mathbf{[2]}}$, formuló la Regla de las Fases, que calcula el número de variables independientes $F$ que se necesitan para determinar el estado de un sistema que consiste en una mezcla de $N$ componentes y $\Pi$ fases que coexisten en equilibrio termodinámico:

$F = N - \Pi + 2 \tag{4}$

Si es el sistema consiste en una sola sustancia ($N=1$) y una sola fase presente (sólido, líquido o vapor), $\Pi=1$, entonces se necesitan $F= 1 - 1 +2 = 2$ variables independientes para determinar el estado del sistema.

En virtud al teorema de Duhem, tenemos que el volumen específico ($V$) es función de $T$ y $P$:

$\displaystyle V = V(T,P) ~~~\Rightarrow~~~ dV = \left ( \frac{\partial V}{\partial T} \right)_{\! P} dT + \left ( \frac{\partial V}{\partial P} \right)_{\! T} dP \tag{5a}$

o para cualquier otra variable $\Psi$, donde esta puede ser cualquier propiedad como la energía interna ($U$), la entalpía ($H$), energía libre de Gibbs ($G$), energía libre de Helmholtz ($A$) o entropía ($S$), tenemos de la misma forma:

$\displaystyle \Psi = \Psi(T,P) ~~~\Rightarrow~~~ d \Psi = \left ( \frac{\partial \Psi}{\partial T} \right)_{\! P} dT + \left ( \frac{\partial \Psi}{\partial P} \right)_{\! T} dP \tag{5b}$

Pero se puede generalizar, no necesariamente $T$ y $P$ deben ser las variables independientes. Un ejemplo común de variables de entrada pueden ser la presión y la entalpía.


Utilizaremos también las Relaciones de Maxwell (demostradas al final de esta respuesta):

$ \left \{ \begin{array}{l} \displaystyle  \left ( \frac{\partial T}{\partial V} \right)_{\! S} = -\left ( \frac{\partial P}{\partial S} \right)_{\! V}   \\ \displaystyle  \left ( \frac{\partial T}{\partial P} \right)_{\! S} = \left ( \frac{\partial V}{\partial S} \right)_{\! P}    \\ \displaystyle  \left ( \frac{\partial V}{\partial T} \right)_{\! P} = -\left ( \frac{\partial S}{\partial P} \right)_{\! T}   \\  \displaystyle  \left ( \frac{\partial P}{\partial T} \right)_{\! V} = \left ( \frac{\partial S}{\partial V} \right)_{\! T}  \\  \end{array} \right . \tag{6}$

El físico James Clerk Maxwell no sólo es conocido por su mayor aporte a la humanidad que son las Ecuaciones de Maxwell del Electromagnetismo, siendo la unificación de la física más importante hasta esta fecha y las ecuaciones principales que nos han llevado al grado tecnológico que tenemos hoy en día. También Maxwell es conocido por sus grandes aportes a la termodinámica clásica como lo son estas hermosas relaciones termodinámicas presentadas en (6) y la Teoría Cinética de Gases, específicamente de la mecánica estadística, donde la distribución de Maxwell-Boltzmann se utiliza para describir la distribución de velocidades en los gases ideales. 

Otras identidades que usaremos como base para esta demostración, son las definiciones de calor específico y sus corolarios:

$ \left \{ \begin{array}{l} \displaystyle C_V :=   \left ( \frac{\partial U}{\partial T} \right)_{\! V}   \\  \displaystyle C_P :=   \left ( \frac{\partial H}{\partial T} \right)_{\! P}   \\ \displaystyle \left ( \frac{\partial S}{\partial T} \right)_{\! V}= \frac{C_V}{T}    \\ \displaystyle \left ( \frac{\partial S}{\partial T} \right)_{\! P}= \frac{C_P}{T}    \end{array} \right . \tag{7} $

Ahora sí, respondamos a la pregunta.

__2. Demostración de la Ecuación (1):__

La Ecuación (1) está escrita en función de la densidad ($\rho$) en lugar del volumen específico ($V$). Sin embargo, las Relaciones de Maxwell (RM) están expresadas en términos de $V$, por lo que partiremos de esta como variable independiente, y al final podemos relacionar ambas mediante $\rho = 1/V$. Empezaremos con $ \left ( \partial P / \partial V \right)_{ S}$ en lugar de $ \left ( \partial P / \partial \rho \right)_{ S}$. 

La primera operación es usar la Relación Cíclica (3d) aplicada a las variables $P$, $V$, $S$, luego sustituir la primera y segunda RM de (6), y aplicar la propiedad de la Ecuación (3b) a la segunda RM:

$\displaystyle \left ( \frac{\partial P}{\partial V} \right)_{\! S} = \underbrace{- \left ( \frac{\partial P}{\partial S} \right)_{\! V} }_{=\text{RM } \#1} \cdot  \underbrace{ \left ( \frac{\partial S}{\partial V} \right)_{\! P} }_{=\text{RM } \#2} = \left ( \frac{\partial T}{\partial V} \right)_{\! S} \left ( \frac{\partial P}{\partial T} \right)_{\! S}  \tag{8}$


Aplicamos la Relación Cíclica a cada una de las derivadas parciales en el lado derecho de la ecuación anterior, es decir al conjunto de variables ($T,V,S$) y ($P,T,S$), respectivamente, luego aplicar la tercera y cuarta relación de Maxwell de (6) y las expresiones de los calores específicos de (7):

$ \left \{ \begin{array}{l} \displaystyle \left ( \frac{\partial T}{\partial V} \right)_{\! S} = - \underbrace{ \left ( \frac{\partial T}{\partial S} \right)_{\! V} }_{=T/C_V} \cdot \underbrace{ \left ( \frac{\partial S}{\partial V} \right)_{\! T} }_{=\text{RM } \# 4} = - \frac{T}{C_V} \left ( \frac{\partial P}{\partial T} \right)_{\! V}     \\  \displaystyle \left ( \frac{\partial P}{\partial T} \right)_{\! S} = \underbrace{ -  \left ( \frac{\partial P}{\partial S} \right)_{\! T} }_{\text{ RM } \# 3} \cdot \underbrace{ \left ( \frac{\partial S}{\partial T} \right)_{\! P} }_{=C_P/T}  = \left ( \frac{\partial T}{\partial V} \right)_{\! P} \frac{C_P}{T}    \end{array} \right . \tag{9} $

Se sustituyen los resultados de (9) dentro de (8) y se aplica la Relación Cíclica por última vez:

$\displaystyle \left ( \frac{\partial P}{\partial V} \right)_{\! S}  = - \frac{T}{C_V} \underbrace{\left ( \frac{\partial P}{\partial T} \right)_{\! V} \left ( \frac{\partial T}{\partial V} \right)_{\! P} }_{ \displaystyle = - \left ( \frac{\partial P}{\partial V} \right)_{\! T} } \frac{C_P}{T} = \frac{C_P}{C_V} \left ( \frac{\partial P}{\partial V} \right)_{\! T}  \tag{10}$

Sabiendo que la densidad es el recíproco del volumen específico, entonces

$\displaystyle \rho = \frac{1}{V} ~~~\Rightarrow~~~ \left ( \frac{\partial \rho}{\partial V} \right)_{\! \Psi} = \frac{\partial }{\partial V} \left ( \frac{1}{V} \right) = -\frac{1}{V^2} \tag{11}$

donde $\Psi$ es cualquier variable termodinámica. Finalmente aplicamos la regla de la cadena (3a) en el lado izquierdo y derecho de la Ecuación (10) y sustituimos (11) en el que se tachan $\left ( \partial \rho / \partial V \right)_{ \Psi}$ en ambos lados de la igualdad:

$\displaystyle \left ( \frac{\partial P}{\partial \rho} \right)_{\! S} \underbrace{ \left ( \frac{\partial \rho}{\partial V} \right)_{\! S} }_{=-1/V^2} =  \frac{C_P}{C_V} \left ( \frac{\partial P}{\partial \rho} \right)_{\! T} \underbrace{ \left ( \frac{\partial \rho}{\partial V} \right)_{\! T} }_{=-1/V^2} ~~~\Rightarrow~~~ \left ( \frac{\partial P}{\partial \rho} \right)_{\! S} = \frac{C_P}{C_V} \left ( \frac{\partial P}{\partial \rho} \right)_{\! T} \tag{12}$

quedando así demostrada la Ecuación (1) y resolviendo esta pregunta.



__3. Demostración de la Relación Cíclica (3c):__

Sean $x = X(y,z)$ y $y = Y(x,z)$ la representación de la misma superficie en $\mathbb{R}^3$ en dos planos diferentes, entonces la diferencial exacta de $x$ y $y$ son respectivamente

$\displaystyle x = X(y,z) ~~~\Rightarrow~~~ dx = \left ( \frac{\partial x}{\partial y} \right)_{\! z} dy + \left ( \frac{\partial x}{\partial z} \right)_{\! y} dz \tag{13a}$

$\displaystyle y = Y(x,z) ~~~\Rightarrow~~~ dy = \left ( \frac{\partial y}{\partial x} \right)_{\! z} dx + \left ( \frac{\partial y}{\partial z} \right)_{\! x} dz \tag{13b}$

Sustituyendo el diferencial $dx$ de (13a) en (13b) y agrupando los términos que acompañan a $dy$ y a $dz$:

$\displaystyle dy = \left ( \frac{\partial y}{\partial x} \right)_{\! z} \left (  \left ( \frac{\partial x}{\partial y} \right)_{\! z} dy + \left ( \frac{\partial x}{\partial z} \right)_{\! y} dz  \right) + \left ( \frac{\partial y}{\partial z} \right)_{\! x} dz \tag*{}$

$\displaystyle  \left ( \left ( \frac{\partial y}{\partial x} \right)_{\! z}  \left ( \frac{\partial x}{\partial y} \right)_{\! z}  - 1 \right) dy + \left ( \left ( \frac{\partial y}{\partial x} \right)_{\! z} \left ( \frac{\partial x}{\partial z} \right)_{\! y}    + \left ( \frac{\partial y}{\partial z} \right)_{\! x} \right ) dz = 0 \tag{14a}$

La Ecuación (14a) siempre se satisface si cada término que acompaña a $dy$ y a $dz$ son iguales a cero, entonces

$\left \{ \begin{array}{l} \displaystyle \left ( \frac{\partial y}{\partial x} \right)_{\! z}  \left ( \frac{\partial x}{\partial y} \right)_{\! z}  - 1 = 0   \\  \displaystyle \left ( \frac{\partial y}{\partial x} \right)_{\! z} \left ( \frac{\partial x}{\partial z} \right)_{\! y}    + \left ( \frac{\partial y}{\partial z} \right)_{\! x} = 0  \end{array} \right. \tag{14b}$

La primera expresión de (14b) es el recíproco de la derivada que es la misma propiedad (ii) escrita en la Ecuación (3b), y la segunda expresión de (14b) es la relación cíclica.


__4. Demostración de las Relaciones de Maxwell:__

Partimos de la Primera Ley de la Termodinámica para un proceso cerrado en forma diferencial:

$dQ - dW = dU \tag{15a}$

donde $dQ$ es el diferencial del calor transferido, $dW$ es el diferencial del trabajo y $dU$ el cambio diferencial de energía interna. Para un proceso reversible, el calor es $dQ=TdS$ y el trabajo es $dW=PdV$, entonces

$dU = TdS - PdV  \tag{15b}$

También de la Primera Ley de la Termodinámica para un proceso abierto, sin cambio de energía interna, en forma diferencial, tenemos que

$dQ - dW = dH    ~~~\Rightarrow~~~ dH = TdS + VdP \tag{16}$

En (16), el trabajo reversible para un sistema abierto se define en forma diferente, este es igual a $dW= - VdP$. De las definiciones de la energía libre de Gibbs ($G$) y energía libre de Helmholtz ($A$):

$G:= H-TS ~~~\Rightarrow~~~ dG = \underbrace{dH - TdS}_{(16)} - SdT ~~~\Rightarrow~~~ dG = - SdT + VdP  \tag{17}$

$A:= U-TS ~~~\Rightarrow~~~ dA = \underbrace{dU - TdS}_{(15b)} - SdT ~~~\Rightarrow~~~ dA = - SdT - PdV  \tag{18}$

Usando el Concepto de Estado y la diferencial exacta (3c) a cada una de las ecuaciones (15b)-(18), tenemos que

$\left \{   \begin{array}{l} \displaystyle  dU = TdS - PdV = \left ( \frac{\partial U}{\partial S} \right)_{\! V} dS + \left ( \frac{\partial U}{\partial V} \right)_{\! S} dV     \\ \displaystyle dH = TdS + VdP = \left ( \frac{\partial H}{\partial S} \right)_{\! P} dS + \left ( \frac{\partial H}{\partial P} \right)_{\! S} dP      \\ \displaystyle  dG = - SdT +  VdP  = \left ( \frac{\partial G}{\partial T} \right)_{\! P} dT + \left ( \frac{\partial G}{\partial P} \right)_{\! T} dP    \\ \displaystyle dA = - SdT - PdV = \left ( \frac{\partial A}{\partial T} \right)_{\! V} dT + \left ( \frac{\partial A}{\partial V} \right)_{\! T} dV     \end{array} \right . \tag{19a}$

Comparando término a término, se obtienen ocho relaciones:

$ \left \{   \begin{array}{cc} \displaystyle    T = \left ( \frac{\partial U}{\partial S} \right)_{\! V},  &  \displaystyle  -P = \left ( \frac{\partial U}{\partial V} \right)_{\! S}     \\ \displaystyle  T = \left ( \frac{\partial H}{\partial S} \right)_{\! P},     &  \displaystyle  V =  \left ( \frac{\partial H}{\partial P} \right)_{\! S}      \\ \displaystyle    -S = \left ( \frac{\partial G}{\partial T} \right)_{\! P} ,  &  \displaystyle   V = \left ( \frac{\partial G}{\partial P} \right)_{\! T}     \\ \displaystyle   -S = \left ( \frac{\partial A}{\partial T} \right)_{\! V} ,  &  \displaystyle  -P = \left ( \frac{\partial A}{\partial V} \right)_{\! T}   \end{array} \right . \tag{19b} $


Usando el teorema de Clairaut-Schwarz a cada línea de la expresión anterior, en el que las segundas derivadas parciales cruzadas son iguales (las funciones tienen que tener segundas derivadas continuas):

$ \left \{   \begin{array}{c} \displaystyle    \left ( \frac{\partial T}{\partial V} \right)_{\! S} =  \frac{\partial^2 U}{\partial V \partial S} = \frac{\partial^2 U}{\partial S \partial V} =  - \left ( \frac{\partial P}{\partial S} \right)_{\! V}     \\ \displaystyle  \left ( \frac{\partial T}{\partial P} \right)_{\! S} =  \frac{\partial^2 H}{\partial P \partial S}  = \frac{\partial^2 H}{\partial S \partial P}= \left ( \frac{\partial V}{\partial S} \right)_{\! P}     \\ \displaystyle    - \left ( \frac{\partial S}{\partial P} \right)_{\! T} =  \frac{\partial^2 G}{\partial P \partial T} = \frac{\partial^2 G}{\partial T \partial P}  =  \left ( \frac{\partial V}{\partial T} \right)_{\! P}     \\ \displaystyle   -\left ( \frac{\partial S}{\partial V} \right)_{\! T} = \frac{\partial^2 A}{\partial V \partial T} = \frac{\partial^2 A}{\partial T \partial V}  = -  \left ( \frac{\partial P}{\partial T} \right)_{\! V} \end{array}  \right . \tag{19c} $

que son las mismas ecuaciones dadas en (6). Esta demostración se puede conseguir fácilmente en cualquier libro de termodinámica básica, por ejemplo el capítulo 12 de Cengel & Boles (2007) $^{\mathbf{[3]}}$ el cual recomiendo que lo lean si están muy interesados en aprender las operaciones algebraicas para manipular relaciones termodinámicas.

__5. Ecuaciones/tablas de Bridgman:__

En caso que se nos dificulte encontrar alguna derivada parcial en particular, podemos ayudarnos con las Ecuaciones de Bridgman $^{\mathbf{[4]}}$ que son una colección de diferenciales para las derivadas parciales. Reescribimos en la derivada, separando cada diferencial en el numerador y denominador en la siguiente manera informal:

$\displaystyle \left ( \frac{\partial Z}{\partial Y} \right)_{\! X} = \frac{ (\partial Z)_X}{(\partial Y)_X} \tag{20}$

y luego sustituimos las expresiones correspondientes a $(\partial Z)_X$ y $(\partial Y)_X$. Se pueden obtener directamente de su página en la Wikipedia, pero tienen la desventaja de que sólo están expresadas en función del calor específico $C_P$. En este caso para la velocidad del sonido, buscamos las relaciones para $(\partial P)_S$ y $(\partial V)_S$

$ \displaystyle \left .  \begin{array}{l} \displaystyle (\partial P)_S = -\frac{C_P}{T}     \\  \displaystyle (\partial V)_S = -\frac{C_P}{T} \left ( \frac{\partial V}{\partial P} \right)_{\! T} - \left ( \frac{\partial V}{\partial T} \right)_{\! P}^2   \end{array}  \right \}   ~~~\Rightarrow~~~ \left ( \frac{\partial P}{\partial V} \right)_{\! S} = \frac{\displaystyle -\frac{C_P}{T}}{\displaystyle -\frac{C_P}{T} \left ( \frac{\partial V}{\partial P} \right)_{\! T} - \left ( \frac{\partial V}{\partial T} \right)_{\! P}^2} \tag{21}$

El denominador de (21) está relacionado con el calor específico a volumen constante, $C_V$:

$ \displaystyle \left .  \begin{array}{l} \displaystyle (\partial U)_V  = T \left (  \frac{C_P}{T} \left ( \frac{\partial V}{\partial P} \right)_{\! T} - \left ( \frac{\partial V}{\partial T} \right)_{\! P}^2 \right )     \\  \displaystyle (\partial T)_V = \left ( \frac{\partial V}{\partial P} \right)_{\! T}   \end{array}  \right \}   ~~~\Rightarrow~~~ \\ \displaystyle C_V:= \left ( \frac{\partial U}{\partial T} \right)_{\! V} = \frac{\displaystyle  T \left (  \frac{C_P}{T} \left ( \frac{\partial V}{\partial P} \right)_{\! T} - \left ( \frac{\partial V}{\partial T} \right)_{\! P}^2 \right )}{\displaystyle \left ( \frac{\partial V}{\partial P} \right)_{\! T} } \tag*{}$

de la expresión anterior deducimos que

$ \displaystyle \frac{C_P}{T} \left ( \frac{\partial V}{\partial P} \right)_{\! T} - \left ( \frac{\partial V}{\partial T} \right)_{\! P}^2  = \frac{C_V}{T} \left ( \frac{\partial V}{\partial P} \right)_{\! T} \tag{22}  $

sustituyendo (22) en (21):

$\displaystyle \left ( \frac{\partial P}{\partial V} \right)_{\! S} = \frac{\displaystyle \frac{C_P}{T}}{\displaystyle \frac{C_V}{T} \left ( \frac{\partial V}{\partial P} \right)_{\! T}} = \frac{C_P}{C_V} \left ( \frac{\partial P}{\partial V} \right)_{\! T} \tag{23}$

__6. Información adicional: cálculo de la velocidad del sonido para el aire a nivel del mar:__

Ahora calculemos la velocidad del sonido para un caso particular: aire seco a nivel del mar, con condiciones $T = 290 \text{ K}$ y $P = 101325 \text{ Pa}$. Suponiendo que el aire se comporta como gas ideal, tenemos que $PV=RT ~~~\Rightarrow~~~ P = \rho R T$, sustituyendo directamente en el lado derecho de la igualdad (1):

$\displaystyle c^2=  \frac{C_P}{C_V}  \left ( \frac{\partial (\rho R  T)}{\partial \rho} \right)_{\! T} \tag*{}$


$\displaystyle c = \sqrt{ \frac{C_P}{C_V}  RT } \tag{24}$

El aire se aproxima como un gas con composición estándar de 21% de oxígeno ($\text{O}_2$) y 79% de nitrógeno ($\text{N}_2$) se puede aproximar como una sustancia pura. El peso molecular del aire es $M = 28.96 \text{ kg/kmol}$. Dado que esta mezcla es de gases diatómicos (moléculas formadas sólo con dos átomos), entonces el $C_V/R$ teórico es $5/2$, por lo que el $C_P/R$ teórico es de $7/2$ (obtenida de la relación entre calores específicos de los gases ideales $C_P - C_V = \bar{R}$), $\bar{R}$ es la constante universal de los gases ideales $8314.5 \text{ J}\cdot \text{kmol}^{-1} \cdot{\text{K}}^{-1}$, y la constante de los gases en unidades másicas se obtiene dividiendo entre el peso molecular, $R = \bar{R}/M$. Tenemos que

$\displaystyle c = \sqrt{ \frac{7/2}{5/2} \cdot \left ( \frac{8314.5 \text{ J}\cdot \text{kmol}^{-1} \cdot{\text{K}}^{-1}}{28.96 \text{ kg/kmol}} \right) \cdot (290 \text{ K}) } = 341.41 \text{ m/s} \tag{25}$

Según la página web de la NASA $^{\mathbf{[5]}}$, dice que la velocidad del sonido en condiciones normales es de $1100 \text{ ft/s} = 335.3 \text{ m/s}$, así que nuestra aproximación con la ecuación de gas ideal tiene una desviación relativa menor al 2%.

__8. Apéndice A. Código utilizado para calcular la velocidad del sonido a nivel del mar:__


```python
!pip install coolprop
import CoolProp

EOS = 'HEOS' # ecuación de estado PVT
             # Helmholtz Equation of State
Fluid = 'Air' # fluido = 'aire'
ASclass = CoolProp.AbstractState(EOS, Fluid)
P0 = 101325  #presión atmosférica [Pa]
             # a nivel del mar
T0 = 290     #temperatura a nivel del mar [K]

# definimos la densidad como una función de la temperatura y presión
def SpeedOfSound(T, P):
    ASclass.update(CoolProp.PT_INPUTS, P, T)
    return ASclass.speed_sound()

print('c = %.2f m/s' % SpeedOfSound(T0, P0))
```

En esta ecuación diferencial no aparece la variable independiente explícitamente, por lo que vamos a asumir que es $x$ para (o puede ser la letra que sea):

$ \displaystyle 2y''=3y^2~~~\Rightarrow~~~2 \frac{d^2y}{dx^2}=3y^2 \tag 1$

La Ecuación (1) es una buena candidata para el cambio de la variable independiente $x \mapsto y $, ya que no aparece la variable independiente cuando se escribe en notación de Lagrange. El cambio de variable es el siguiente:

$\displaystyle z=\frac{dy}{dx} \tag{2a}$

Derivando (2a) respecto a $x$ y aplicando la regla de la cadena en una variable en el lado izquierdo de la igualdad:

$\displaystyle \frac{dz}{dx}=\frac{d^2y}{dx^2}~~~\Rightarrow~~~\frac{dz}{dy} \cdot \underbrace{\frac{dy}{dx}}_{=z} = \frac{d^2y}{dx^2} \\ \displaystyle \frac{d^2y}{dx^2}=z\frac{dz}{dy} \tag{2b} $

Sustituyendo (2b) en (1), transformamos a una ecuación diferencial de primer orden con $y$ como variable independiente:

$ \displaystyle 2z\frac{dz}{dy}=3y^2 \tag 3$

La Ecuación (3) es sencilla de resolver aplicando separación de variables e integrando ambos lados. La integración la podemos hacer de forma indefinida (sin poner límites de integración) pero añadiendo una constante:

$ \displaystyle 2z\,dz=3y^2 \, dy ~~~\Rightarrow~~~ 2 \int z\,dz=3 \int y^2 \, dy +C_1\tag*{}$

por lo tanto

$ \displaystyle 2 \cdot \frac{z^2}{2}=3 \cdot \frac{y^3}{3}+C_1~~~\Rightarrow~~~z^2=y^3+C_1 \tag 4$

deshaciendo el cambio de variable, $z=dy/dx$:

$ \left ( \dfrac{dy}{dx} \right )^2 =y^3+C_1 ~~~\Rightarrow~~~ \dfrac{dy}{dx}= \pm \sqrt{y^3+C_1} \tag{5}$

Separando variables e integrando (y nuevamente añadiendo otra constante):

$\displaystyle \int \frac{dy}{\sqrt{y^3+C_1}}= \pm x + C_2 \tag 6 $

La integral del lado izquierdo de la igualdad en (6) no tiene antiderivada elemental, por lo que debemos dejarla expresada. El signo $\pm$ indica que ambos signos (el positivo y el negativo) son válidos. Hay una restricción en el dominio, que lo que esté adentro de la raíz cuadrada debe ser positivo:

$y^3+C_1>0 \tag 7$

Esta sería la respuesta a esta pregunta ya que esta no especifica dos condiciones iniciales o de borde. Los datos para resolverla no necesariamente deben ser los típicos para estas ecuaciones diferenciales ordinarias, e.g. $y(0)=y_0, y'(0)=y'_0$, pueden ser dos condiciones de borde para $y(x)$ y $y'(x)$ en un intervalo $a \le x \le b$, e incluso el límite superior puede ser infinito. También, podríamos graficar algunas curvas de nivel dándole valores a las constantes $C_1$ y $C_2$, por supuesto integrando numéricamente. Sin embargo aquí les propongo otra cosa más interesante que es darle interpretación física a esta ecuación y la terminamos de resolver analíticamente (e incluso numéricamente con algún método built-in de alguna librería popular de código abierto), además, preferiblemente la resolvemos con la restricción que sea acotada $|y(x)| \le y_{\text{max}}<\infty$.

Una posible aplicación directa de la ecuación diferencial (1), es el __transporte por difusión de un soluto A que sufre una reacción química irreversible en un medio semi-infinito__ (en coordenadas rectangulares) en todo su volumen que puede ser un medio acuoso $\mbox{A}_{\text{(ac)}} \xrightarrow[]{~~~~k_V~~~~} \mbox{B}_{\text{(ac)}}$, siendo de cinética de segundo orden (la tasa de consumo es proporcional al cuadrado de la concentración del reactivo A), en estado estacionario (la concentración no varía con el tiempo), flujo unidireccional e isotérmico. El soluto A se difunde desde la interfase plana en $x=0$ con una concentración fija $C_A=C_{A0}$, tal como se muestra en la siguiente figura. El objetivo es encontrar el perfil de concentraciones de A, ($C_A$), en función de la variable espacial $x$.

El balance de moles del soluto A viene dada por la siguiente Ecuación (8):

$ \displaystyle \frac{\partial C_A}{\partial t} + \nabla \cdot \mathbf{N}_A = R_{A,V} \tag 8$

donde $\nabla \cdot (~\cdot~)$ es el operador diferencial divergencia, $\mathbf{N}_A$ es el vector flujo molar del soluto A por unidad de área, y $R_{A,V}$ es la velocidad de reacción química por unidad de volumen, que es negativo para los reactantes (ya que se consumen) y positivo para los productos (ya que se producen), donde todas estas velocidades están relacionadas por la misma estequiometría de la reacción.


La ley de Fick es la ecuación más importante que rige el flujo por difusión de una molécula en un medio, esta viene dada en forma vectorial mediante

$ \displaystyle \mathbf{N}_A = - \mathcal{D}_{Ae} \nabla C_A \tag 9$

donde $\mathcal{D}_{Ae}$ es el coeficiente de difusión efectivo de A en el medio (tiene unidades de $\text{longitud}^2 \times \text{tiempo}^{-1}$, que depende de la temperatura, presión y  composición de la mezcla, pero aquí asumimos que es independiente de la concentración, ya que generalmente el soluto A se encuentra diluido en el solvente. Simplificando para este problema en particular, y sustituyendo la Ecuación (9) dentro de la Ecuación (8) tenemos que

$  \left \{ \begin{array}{l} \displaystyle \underbrace{\frac{\partial C_A}{\partial t}}_{=0} + \underbrace{\nabla \cdot \mathbf{N}_A}_{ =\partial N_{Ax} / \partial x} = \underbrace{R_{A,V}}_{=-k_V C_{A}^2} \\ \displaystyle \underbrace{\mathbf{N}_A}_{=N_{Ax} \mathbf{\hat{i}}} = - \mathcal{D}_{Ae} \underbrace{ \nabla C_A}_{=\frac{\partial C_A}{\partial x} \mathbf{\hat{i}}}  \end{array}  \right . ~~~~~ \Rightarrow ~~~~ \displaystyle \frac{d}{dx} \left ( -\mathcal{D}_{Ae} \frac{dC_A}{dx} \right) = -k_V C_A^2 \tag{10a}$

Reescribiendo (10a) y definiendo la siguiente variable dependiente $y = C_A/C_{A0}$ que es adimensional:

$\displaystyle \frac{d^2 C_A}{dx^2}=\frac{k_V}{\mathcal{D}_{Ae}}C_A^2 ~~~~\Rightarrow ~~~~ \frac{d^2 y}{dx^2}=\frac{k_V C_{A0}}{\mathcal{D}_{Ae}}y^2 \tag{10b}$

Aquí en este problema, $k_V C_{A0} / \mathcal{D}_{Ae} = 3/2$. El problema de valor frontera está definido por

$\displaystyle \left \{ \begin{array}{lcc} \displaystyle y''= \frac{3}{2} y^2 & \mbox{si} & 0 \le x < \infty \\
          y(0) = 1 & & \\ \displaystyle \lim_{x \to \infty} \frac{dy}{dx} = 0 & & \end{array} \right . \tag{11}$
          
Aplicando la condicion de borde en el infinito ($x \to \infty$), el cual no hay concentración del soluto A en el infinito ($y(x \to \infty) = 0$) y no hay flujo ($y'(x \to \infty) = 0$), en la Ecuación (5), obtenemos el valor de la constante $C_1$:

$ \displaystyle \left ( \lim_{x \to \infty} \dfrac{dy}{dx} \right )^2 =\left ( \lim_{x \to \infty} y \right )^3+C_1 ~~~ \Rightarrow ~~~ 0^2 = 0^3 + C_1  ~~~ \Rightarrow ~~~ C_1 = 0 \tag{12}$

Por lo que podemos integrar analíticamente la Ecuación (6):

$\displaystyle \int \frac{dy}{\sqrt{y^3+0}}= \pm x + C_2 ~~~\Rightarrow ~~~ \frac{y^{-1/2}}{-1/2} = \pm x +C_2 \tag*{}$

$\displaystyle \frac{1}{\sqrt{y}} = -\frac{1}{2} (\pm x + C_2) \tag{13}$

Aplicando la primera condición de borde $y(0)=1$ en la expresión anterior:

$\displaystyle \frac{1}{\sqrt{ \underbrace{y(0)}_{=1}}} = -\frac{1}{2} (\pm 0 + C_2) ~~~\Rightarrow ~~~ 1=-\frac{1}{2} C_2 ~~~ \Rightarrow ~~~ C_2 = -2\tag{14}$

La Ecuación (13) queda

$\displaystyle \frac{1}{\sqrt{y}} = -\frac{1}{2} (\pm x - 2) \tag{15}$

El lado izquierdo de la Ecuación (15) es positivo, por lo que el lado derecho de esta también debe ser positivo para todo $0 \le x < \infty$, así que seleccionamos el signo menos ($-$) en $\pm$ para que esta sea consistente con una solución real en este intervalo de valores de $x$, entonces:

$\displaystyle \frac{1}{\sqrt{y}} = -\frac{1}{2} (-x - 2) ~~~ \Rightarrow ~~~ \sqrt{y} = \frac{1}{\displaystyle \frac{x}{2} + 1} \tag*{}$

Finalmente tenemos una función que es la solución para este problema en particular:

$\displaystyle y = \frac{1}{\left ( \displaystyle \frac{x}{2} + 1 \right )^2} \tag{16}$

que es una solución acotada para todo $0 \le x < \infty$. Podemos verificar fácilmente que se satisfacen $y(0) = 1, y(x \to \infty) = 0, y'(x \to \infty) = 0$. Si derivamos dos veces, obtenemos la Ecuación (1), tal como se muestra a continuación:

$\displaystyle \frac{dy}{dx} = -2\frac{1}{\left ( \dfrac{x}{2} + 1 \right )^3 } \cdot \frac{1}{2} = -\frac{1}{\left ( \dfrac{x}{2} + 1 \right )^3 }  \tag*{}$

$\displaystyle \frac{d^2y}{dx^2} = -(-3) \frac{1}{\left ( \dfrac{x}{2} + 1 \right )^4} \cdot \frac{1}{2} = \frac{3}{2} \cdot \underbrace{ \left (  \frac{1}{\left ( \dfrac{x}{2} + 1 \right )^2}  \right )^2 }_{=y^2} = \frac{3}{2} y^2  \tag*{}$


Podemos resolver el problema de valor frontera o _boundary value problem (BVP)_ con un pequeño programa en python mediante la función ```scipy.integrate.solve_bvp()``` donde aplica el método de diferencias finitas. Para esto, primero debemos separar la Ecuación (1), que es de segundo orden, en dos ecuaciones diferenciales de primer orden cada una. Definimos $Y_0 = y, Y_1 = dy/dx$, por lo tanto:

$ \displaystyle \left \{ \begin{array}{lcc} \dfrac{d Y_0}{dx} = Y_1 & \mbox{si} & 0 \le x \le \ell \\ \dfrac{d Y_1}{dx} = \dfrac{3}{2} Y_0^2 & \mbox{si} & 0 \le x \le \ell \end{array} \right . \tag{17} $

Las condiciones de borde igualadas a cero vienen dadas por

$ \displaystyle \left \{ \begin{array}{c} Y_0(x=0) - 1 = 0 \\ Y_1(x=\ell) - 0 =0 \end{array} \right . \tag{18} $

Aquí no podemos introducir el infinito como condición de borde, por lo que utilizamos un valor de $x = \ell$ muy grande, por ejemplo $\ell = 1000$, ya que la derivada evaluada en este valor es muy cercana a cero. Debemos utilizar un número de nodos lo suficientemente grande para que no cause oscilaciones debido al error por truncación en diferencias finitas, en este caso $N_x = 1001$, el cual fija un tamaño de paso de $\Delta x = 1$. Graficamos $y = C_A/C_{A0}$ en función de $x$ para el intervalo $0 \le x \le 10$ y la comparamos con la solución analítica obtenida anteriormente calculando la desviación relativa absoluta promedio y la desviación máxima entre ambas curvas:


````python
import scipy as sp
import numpy as np
import matplotlib.pyplot as plt

Nx = 1001
Lmax_integration = 1000
Lmax_plot = 10

def fun(x, Y):     
    return np.vstack((Y[1], 3/2*Y[0]**2))

def boundary_cond(ya, yb):
    return np.array([ya[0] - 1, yb[1] - 0])

x = np.linspace(0, Lmax_integration, Nx)
y = np.zeros((2, x.size))

sol = sp.integrate.solve_bvp(fun, boundary_cond, x, y)

x_plot = np.linspace(0, Lmax_plot, Nx)
y_approx = sol.sol(x_plot)[0]
y_exact = 1/(x_plot/2 + 1)**2

plt.plot(x_plot, y_approx, 'k--', lw=2)
plt.plot(x_plot, y_exact, 'k-', lw=2)
plt.legend(['Numérico','Exacto'])
plt.xlabel("Posición $x$")
plt.ylabel("Concentración adimensional, $y = C_A/C_{A0}$")
plt.xlim(0, Lmax_plot)
plt.ylim(0, 1)
plt.grid()
plt.show()
relative_deviation = np.abs(y_approx-y_exact)/y_exact
percent_AARD = 100*np.mean(relative_deviation)
percent_max_ARD = 100*np.max(relative_deviation)

print('Abs Avg relative deviation = %.6f percent' % percent_AARD)
print('Max Abs relative deviation = %.6f percent' % percent_max_ARD)
````

Tal como podemos ver en la gráfica anterior, no hay mucha diferencia entre la solución $y(x)$ exacta y la numérica, ya que difieren en una desviación absoluta promedio menor al 1%.

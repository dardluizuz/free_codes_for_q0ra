Para obtener el valor de esta suma podríamos usar Series de Fourier, una herramienta muy importante en el análisis de señales $^{\mathbf{[1]}}$. Sea $f:[-T/2,T/2] \rightarrow \mathbb{R}$ una función $T$-periódica (no es necesario que sea continua), entonces su representación en serie de Fourier viene dada por $^{\mathbf{[2]}}$

$\displaystyle f(t) \sim S_f(t) = \frac{a_0}{2} +  \sum_{n=1}^{\infty} \left ( a_n \cos \left ( \frac{2 \pi n}{T} t \right ) +  b_n \sin \left ( \frac{2 \pi n}{T} t \right )   \right ) \tag{1}$

donde los coeficientes de la serie se obtienen con la propiedad de ortogonalidad de las funciones coseno y seno:

$\displaystyle a_n = \frac{\left \langle f(t), \cos \left ( \frac{2 \pi n}{T} t \right )  \right \rangle}{\left \langle \cos \left ( \frac{2 \pi n}{T} t \right ), \cos \left ( \frac{2 \pi n}{T} t \right ) \right  \rangle} = \frac{2}{T} \int_{-T/2}^{T/2} f(t) \cos \left ( \frac{2 \pi n}{T} t \right ) \, dt \tag{2a}$

$\displaystyle b_n = \frac{\left \langle f(t), \sin \left ( \frac{2 \pi n}{T} t \right )  \right \rangle}{\left \langle \sin \left ( \frac{2 \pi n}{T} t \right ), \sin \left ( \frac{2 \pi n}{T} t \right ) \right  \rangle} = \frac{2}{T} \int_{-T/2}^{T/2} f(t) \sin \left ( \frac{2 \pi n}{T} t \right ) \, dt \tag{2b}$

El coeficiente del término independiente se encuentra sustituyendo $n=0$ en (2a):

$\displaystyle a_0 = \frac{2}{T} \int_{-T/2}^{T/2} f(t)  \, dt \tag{2c}$

La serie de Fourier de $f$ converge en el mismo valor que la función $f$ si es continua en el punto $t = t_0$, pero en caso que la función $f$ sea discontinua en dicho punto, entonces su serie de Fourier converge en el punto medio del salto discontinuo:

$\displaystyle \lim_{t \to t_0} S_f(t) = \frac{\displaystyle \lim_{t \to t_0^-} f(t) + \lim_{t \to t_0^+} f(t)}{2} \tag{3}$

donde $t_0^-$ es la izquierda y $t_0^+$ la derecha del punto $t_0$. Esto se conoce como teorema de convergencia puntual de las serie de Fourier o condición de Hölder $^{\mathbf{[3]}}$.

Para calcular el valor de esta suma, tendremos que buscar la serie de Fourier de alguna función y evaluarla en un punto para obtener el valor de la serie. Podríamos ensayar con varios polinomios. Aquí usaremos la siguiente función $2\pi$-periódica $f:[-\pi,\pi] \rightarrow \mathbb{R}$ dada por

$f(t) = \left \{ \begin{array}{lcc} t(\pi + t) & \text{si} & -\pi \le t < 0  \\ t(\pi - t) & \text{si} & 0 \le t \le \pi \end{array} \right . \tag{4}$

compuesta por dos arcos parabólicos. El valor del período es irrelevante, podríamos haber usado otro, pero usamos $T = 2 \pi$ por conveniencia, y sin pérdida de generalidad ya que el resultado será el mismo. La gráfica de la función dada en la Ecuación (4) se muestra a continuación:

Como podemos observar en la Figura 1, $f(t)$ es una función IMPAR, es decir, satisface $f(-t) = - f(t)$, por lo que los coeficientes $a_n$ y el término independiente $a_0$ valen cero, sólo calculamos los coeficientes $b_n$. Utilizando la Ecuación (2b) y dado que el producto de dos funciones impares es una función par usamos la propiedad de simetría $\int_{-a}^{a} f(t) \,dt = 2 \int_0^a f(t) \, dt$:

$\displaystyle b_n = \frac{2}{2 \pi} \int_{-\pi}^{\pi} f(t) \sin \left ( \frac{2 \pi n}{2\pi} t \right ) \, dt = \frac{2}{ \pi} \int_{0}^{\pi} f(t) \sin (nt) \, dt \tag{5a}$

A continuación trabajamos con pura "carpintería" para calcular la integral, aplicamos integración por partes tres veces:

$\displaystyle \int_0^{\pi} f(t) \sin (nt) \, dt = \left . f(t) \left ( - \frac{\cos (nt)}{n}  \right ) \right |_{\, 0}^{\, \pi} + \frac{1}{n} \int_0^{\pi} f'(t) \cos(nt) \, dt = \\ \displaystyle \frac{f(0) \cos(0) - f(\pi) \cos(\pi n)}{n} + \frac{1}{n} \int_0^{\pi} f'(t) \cos(nt) \, dt  \tag{5b}$

$\displaystyle \int_0^{\pi} f'(t) \cos(nt) \, dt = \left . f'(t) \left ( \frac{\sin (nt)}{n}  \right ) \right |_{\, 0}^{\, \pi} - \frac{1}{n} \int_0^{\pi} f''(t) \sin(nt) \, dt = \\ \displaystyle \underbrace{\frac{f'(\pi) \sin(\pi n) - f'(0) \sin(0) }{n}}_{ = 0} - \frac{1}{n} \int_0^{\pi} f''(t) \sin(nt) \, dt \tag{5c}$

$\displaystyle \int_0^{\pi} f''(t) \sin (nt) \, dt = \left . f''(t) \left ( - \frac{\cos (nt)}{n}  \right ) \right |_{\, 0}^{\, \pi} + \frac{1}{n} \int_0^{\pi} f'''(t) \cos(nt) \, dt = \\ \displaystyle \frac{f''(0) \cos(0) - f''(\pi) \cos(\pi n)}{n} + \frac{1}{n} \int_0^{\pi} f'''(t) \cos(nt) \, dt \tag{5d}$

De la gráfica de $f$ tenemos que $f(0) = f(\pi) = 0$. Las primeras tres derivadas de $f$ en $0 \le t \le \pi$ son

$f(t) = t(\pi - t) = \pi t - t^2 ~~~\Rightarrow ~~~ f'(t) = \pi - 2t ~~~\Rightarrow ~~~ f''(t) = -2 ~~~\Rightarrow ~~~ f'''(t) = 0 \tag{6} $

Por lo que $f''(0) = f'''(\pi) = -2$ y $f'''(t)=0$ en todo el intervalo $0 \le t \le \pi$. Sustituyendo estos resultados en (5d):

$\displaystyle \int_0^{\pi} f''(t) \sin (nt) \, dt = \displaystyle \frac{(-2) \cdot 1 - (-2) \cos(\pi n)}{n} + \frac{1}{n} \underbrace{\int_0^{\pi} f'''(t) \cos(nt) \, dt}_{=0} = \\ \displaystyle  -2 \cdot \frac{1 - \cos(\pi n)}{n} \tag{7a}$

Sustituyendo (7a) en (5c)

$\displaystyle \int_0^{\pi} f'(t) \cos(nt) \, dt = - \frac{1}{n} \int_0^{\pi} f''(t) \sin(nt) \, dt = \\ \displaystyle - \frac{1}{n} \cdot (-2) \cdot \frac{1 - \cos(\pi n)}{n} = \frac{2}{n^2}(1 - \cos(\pi n)) \tag{7b}$

Sustituyendo (7b) en (5b):

$\displaystyle \int_0^{\pi} f(t) \sin (nt) \, dt =  \frac{0 \cdot 1 - 0 \cdot \cos(\pi n)}{n} + \frac{1}{n} \cdot \frac{2}{n^2}(1 - \cos(\pi n)) = \\ \displaystyle \frac{2}{n^3}\underbrace{(1 - \cos(\pi n))}_{\displaystyle  = 2 \sin^2 \left ( \frac{\pi n}{2} \right)} = \frac{4}{n^3} \sin^2 \left ( \frac{\pi n}{2} \right)    \tag{7c}$

Finalmente, sustituyendo el resultado (7c) en (5a) se obtiene una expresión explícita para los $b_n$:

$\displaystyle b_n =  \frac{2}{ \pi} \cdot \frac{4}{n^3} \sin^2 \left ( \frac{\pi n}{2} \right) = \frac{8}{\pi n^3} \sin^2 \left ( \frac{\pi n}{2} \right)  \tag{7d}$

Por lo que la representación de $f$ en serie de Fourier viene dada por

$\displaystyle f(t) \sim S_f(t) = \sum_{n=1}^{\infty} \frac{8}{\pi n^3} \sin^2 \left ( \frac{\pi n}{2} \right) \sin (n t) \tag{8}$

Evaluando $f(t)$ y su respectiva serie de Fourier en el punto $t = \pi/2$ usando las ecuaciones (4) y (8), respectivamente. Los valores $f(\pi/2)$ y $S_f(\pi/2)$ coinciden porque $f$ es una función continua en $t = \pi/2$, en virtud al teorema de convergencia puntual:

$\displaystyle f \left ( \frac{\pi}{2} \right ) = S_f  \left ( \frac{\pi}{2} \right ) ~~~\Rightarrow ~~~ \frac{\pi}{2} \left ( \pi - \frac{\pi}{2}\right ) = \sum_{n=1}^{\infty} \frac{8}{\pi n^3} \sin^2 \left ( \frac{\pi n}{2} \right) \sin \left ( n \cdot \frac{\pi}{2} \right )  \tag*{}$

$\displaystyle \frac{\pi^2}{4} = \sum_{n=1}^{\infty} \frac{8}{\pi n^3} \sin^3 \left ( \frac{\pi n}{2} \right) ~~~\Rightarrow ~~~   \sum_{n=1}^{\infty} \frac{\displaystyle \sin^3 \left ( \frac{\pi n}{2} \right)}{n^3}  = \frac{\pi^3}{32}   \tag{9a}$

Los términos pares de la suma (9a) son cero ya que $\sin ( \pi n) = 0$ y los impares tienen signos alternantes:

$\displaystyle \sum_{n=1}^{\infty} \frac{\displaystyle \sin^3 \left ( \frac{\pi n}{2} \right)}{n^3} = \frac{1}{1^3} + 0 + \frac{(-1)}{3^3} + 0 + \frac{1}{5^3} + \cdots     = \frac{\pi^3}{32} \tag{9b}$

por lo que esta suma es equivalente a

$ \displaystyle \sum_{n=0}^{\infty} \frac{(-1)^n}{(2n + 1)^3}  = \frac{\pi^3}{32}   \tag{9c}$

Sólo falta mostrar que la suma escrita en la pregunta es el doble que la suma (9a):

$\displaystyle \sum_{n= -\infty}^{\infty} \frac{(-1)^n}{(2n + 1)^3} = \sum_{n= -\infty}^{-1} \frac{(-1)^n}{(2n + 1)^3} + \sum_{n=0}^{\infty} \frac{(-1)^n}{(2n + 1)^3} ~~~\Rightarrow ~~~ \\ \displaystyle \sum_{n= \infty}^{1} \frac{(-1)^{-n}}{(-2n + 1)^3} + \sum_{n=0}^{\infty} \frac{(-1)^n}{(2n + 1)^3} = \underbrace{\sum_{n= 1}^{\infty} \frac{(-1)^{(-n + 1)}}{(2n - 1)^3} }_{\displaystyle =\sum_{n=0}^{\infty} \frac{(-1)^n}{(2n + 1)^3}} + \sum_{n=0}^{\infty} \frac{(-1)^n}{(2n + 1)^3} \tag*{}$

por lo tanto

$\displaystyle \sum_{n= -\infty}^{\infty} \frac{(-1)^n}{(2n + 1)^3} = 2 \sum_{n=0}^{\infty} \frac{(-1)^n}{(2n + 1)^3} = 2 \cdot \frac{\pi^3}{32} = \frac{\pi^3}{16} \tag{10}$

__Información adicional:__

Aprovechando la expresión de la serie de Fourier de $f(t)$ dada en (8), podemos usar la Igualdad de Parseval para encontrar el valor de la siguiente suma 

$\displaystyle \sum_{n=0}^{\infty} \frac{1}{(2n + 1)^6}  \tag{11}$

La igualdad de Parseval $^{\mathbf{[2]}}$ viene dada por

$\displaystyle \frac{2}{T} \int_{-T/2}^{T/2} |f(t)|^2 \, dt = \frac{|a_0|^2}{2} + \sum_{n=1}^{\infty} \left ( |a_n|^2 + |b_n|^2 \right) \tag{12}$

Sustituyendo $T = 2 \pi$, $f(t)$ dada en (4) y $b_n$ dada en (7d). Dado que $f \in L^2[\pi,\pi]$ (la integral de su cuadrado es finito en dicho intervalo) entonces esta suma existe:

$\displaystyle 2 \cdot \frac{2}{2 \pi} \int_{0}^{\pi} |t(\pi - t)|^2 \, dt = \frac{|0|^2}{2} + \sum_{n=1}^{\infty} \left ( |0|^2 + \left | \frac{8}{\pi n^3} \sin^2 \left ( \frac{\pi n}{2} \right) \right |^2 \right) \tag*{}$

$\displaystyle  \frac{2}{ \pi} \int_{0}^{\pi} t^2(\pi - t)^2 \, dt = \sum_{n=1}^{\infty}   \frac{64}{\pi^2 n^6} \sin^4 \left ( \frac{\pi n}{2} \right )   \tag*{}$

$\displaystyle  \sum_{n=0}^{\infty}   \frac{1}{(2n + 1)^6} = \frac{\pi^2}{64} \cdot \frac{2}{ \pi} \int_{0}^{\pi} t^2(\pi - t)^2 \, dt = \\ \displaystyle  \frac{\pi^2}{64} \cdot \frac{2}{ \pi} \cdot \pi^5 \underbrace{\int_{0}^{1} \zeta^2(1 - \zeta)^2 \, d \zeta}_{= 1/30} = \frac{\pi^2}{64} \cdot \frac{2}{ \pi} \cdot \pi^5 \cdot \frac{1}{30} = \frac{\pi^6}{960}    \tag{13}$
